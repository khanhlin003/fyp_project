```python
# Extract dividend data from yfinance and merge into official ETF metadata
import pandas as pd
import ast
from datetime import datetime

# Load both datasets
yfinance_metadata = pd.read_csv("/Users/linhtrankhanh/Downloads/fyp/project/data/raw/us market/yfinance_us_etf_metadata.csv")
etf_metadata = pd.read_csv("/Users/linhtrankhanh/Downloads/fyp/project/data/official/us/etf_metadata.csv")

print(f"📊 Loaded {len(yfinance_metadata)} records from yfinance")
print(f"📊 Loaded {len(etf_metadata)} records from official metadata")

# Extract dividend info from yfinance
dividend_data = []

for _, row in yfinance_metadata.iterrows():
    ticker = row['ticker']
    
    try:
        # Parse the info string into a dictionary
        info_dict = ast.literal_eval(row['info']) if isinstance(row['info'], str) else row['info']
        
        if info_dict and isinstance(info_dict, dict):
            # Extract dividend fields
            dividend_record = {
                'ticker': ticker,
                'annual_dividend_rate': info_dict.get('trailingAnnualDividendRate', ''),
                'annual_dividend_yield': info_dict.get('dividendYield', '') or info_dict.get('yield', ''),
                'dividend': info_dict.get('lastDividendValue', ''),  # Last dividend payment
                'dividend_date': ''  # Will be populated if available
            }
            
            # Convert dividend date if available (Unix timestamp to date string)
            if 'lastDividendDate' in info_dict and info_dict['lastDividendDate']:
                try:
                    dividend_record['dividend_date'] = datetime.fromtimestamp(
                        info_dict['lastDividendDate']
                    ).strftime('%Y-%m-%d')
                except:
                    pass
            elif 'exDividendDate' in info_dict and info_dict['exDividendDate']:
                try:
                    dividend_record['dividend_date'] = datetime.fromtimestamp(
                        info_dict['exDividendDate']
                    ).strftime('%Y-%m-%d')
                except:
                    pass
            
            dividend_data.append(dividend_record)
        else:
            print(f"⚠️ No info for {ticker}")
            
    except Exception as e:
        print(f"❌ Error parsing {ticker}: {str(e)[:50]}")
        dividend_data.append({
            'ticker': ticker,
            'annual_dividend_rate': '',
            'annual_dividend_yield': '',
            'dividend': '',
            'dividend_date': ''
        })

# Create DataFrame from extracted dividend data
dividend_df = pd.DataFrame(dividend_data)

print(f"\n✅ Extracted dividend data for {len(dividend_df)} ETFs")
print(f"\n📊 Sample dividend data:")
print(dividend_df.head())

# Merge dividend data into official ETF metadata
# First, drop the empty dividend columns from etf_metadata
etf_metadata_updated = etf_metadata.drop(
    columns=['annual_dividend_rate', 'annual_dividend_yield', 'dividend', 'dividend_date']
)

# Merge with dividend data
etf_metadata_updated = etf_metadata_updated.merge(
    dividend_df,
    on='ticker',
    how='left'
)

# Reorder columns to match original schema
column_order = [
    'ticker', 'etf_name', 'issuer', 'brand', 'inception', 'category', 'asset_class', 'asset_class_size',
    'asset_class_style', 'general_region', 'specific_region', 'expense_ratio', 'aum',
    'index_tracked', '1_month_return', '3_month_return', 'ytd_return', '1_year_return',     
    '3_year_return', '5_year_return', '52_week_low', '52_week_high', 'beta',
    'standard_deviation', 'annual_dividend_rate', 'annual_dividend_yield',
    'dividend', 'dividend_date', 'top_holdings', 'sector_breakdown', 'source',
    'last_updated'
]

etf_metadata_updated = etf_metadata_updated[column_order]

# Save updated metadata
# output_path = "/Users/linhtrankhanh/Downloads/fyp/project/data/official/us/etf_metadata.csv"
# etf_metadata_updated.to_csv(output_path, index=False)

# print(f"\n💾 Saved updated metadata to: {output_path}")
print(f"\n📊 Dividend data summary:")
print(f"   ETFs with dividend rate: {etf_metadata_updated['annual_dividend_rate'].notna().sum()}")
print(f"   ETFs with dividend yield: {etf_metadata_updated['annual_dividend_yield'].notna().sum()}")
print(f"   ETFs with dividend date: {(etf_metadata_updated['dividend_date'] != '').sum()}")

print(f"\n✅ Sample updated data:")
print(etf_metadata_updated[['ticker', 'etf_name', 'annual_dividend_rate', 'annual_dividend_yield', 'dividend_date']].head(10))
```

```python
# still those information 
"""
1. Identifiers & Basic Info
ticker - Symbol (e.g., "IVV")
etf_name - Full name
issuer - Company (BlackRock, Vanguard, etc.)
brand - Sub-brand (iShares, SPDR, etc.)
inception - Launch date
2. Classification (for filtering/discovery)
category - Large Cap Growth Equities, etc.
asset_class - Equity, Fixed Income, etc.
asset_class_size - Large-Cap, Mid-Cap, Small-Cap
asset_class_style - Growth, Value, Blend
general_region - North America, Europe, etc.
specific_region - U.S., Japan, etc.
3. Financial Metrics (for recommendations)
expense_ratio - Annual fee %
aum - Assets Under Management
index_tracked - Benchmark index
4. Performance (for ranking)
1_month_return
3_month_return
ytd_return
1_year_return
3_year_return
5_year_return
5. Risk Metrics
52_week_low
52_week_high
beta - From technicals.volatility
standard_deviation - From technicals.volatility
6. Dividends (for income investors)
annual_dividend_rate
annual_dividend_yield
dividend - Last payment
dividend_date - Last payment date
7. Holdings (for AI chatbot explanations)
top_holdings - Top 10-15 stocks (symbol:share%)
Sector breakdown from holdings_analysis (e.g., "Electronic Technology: 25.1%")
"""

# Prioritize sg_etfs_list data, then fallback to yfinance

import yfinance as yf
import pandas as pd
from datetime import datetime
from tqdm import tqdm
import time

def safe_get(d, key, default=""):
    """Safely get value from dict, return default if missing or None"""
    return d.get(key, default) if d.get(key) is not None else default

def unix_to_date(ts):
    """Convert Unix timestamp to ISO date string"""
    if ts in [None, "", 0]: 
        return ""
    try:
        return datetime.utcfromtimestamp(ts).date().isoformat()
    except:
        return ""

def safe_get_from_row(row, key, default=""):
    """Safely get value from DataFrame row, return default if missing, NaN, or empty"""
    try:
        val = row.get(key, default)
        if pd.isna(val) or val == "":
            return default
        return val
    except:
        return default

# Initialize storage
etf_metadata_list = []
error_tickers = []

print(f"Starting metadata collection for {len(sg_etfs_tickers)} SG ETFs...")
print("=" * 60)

for yf_ticker in tqdm(sg_etfs_tickers, desc="Fetching metadata"):
    try:
        # Extract base ticker (remove .SI suffix)
        base_ticker = yf_ticker.replace('.SI', '')
        
        # Match row from sg_etfs_list using base ticker
        matching_rows = sg_etfs_list[sg_etfs_list["Trading Code"] == base_ticker]
        
        if matching_rows.empty:
            print(f"⚠️ {base_ticker} not found in sg_etfs_list, skipping...")
            error_tickers.append((base_ticker, "Not found in source list"))
            continue
        
        row = matching_rows.iloc[0]
        
        # Fetch yfinance data as fallback
        yf_obj = yf.Ticker(yf_ticker)
        info = yf_obj.info

        # Holdings and sectors (may be empty or fail)
        try:
            holdings = yf_obj.get_holdings()
            if holdings is not None and not holdings.empty:
                holdings_data = holdings.head(10).to_dict("records")
            else:
                holdings_data = []
        except:
            holdings_data = []

        try:
            sectors = yf_obj.get_funds_sector_weightings()
            sector_data = sectors[0] if sectors else {}
        except:
            sector_data = {}

        # Build metadata dictionary - PRIORITIZE sg_etfs_list FIRST
        data = {
            # 1. Identifiers
            "ticker": base_ticker,
            "etf_name": safe_get_from_row(row, "Trading Name") or safe_get(info, "longName", ""),
            "issuer": safe_get_from_row(row, "Fund Manager") or safe_get(info, "fundFamily", ""),
            "brand": safe_get(info, "fundFamily", ""),  # Not in sg_etfs_list, use yfinance
            "inception": unix_to_date(safe_get(info, "fundInceptionDate")),  # Not in sg_etfs_list

            # 2. Classification
            "category": safe_get_from_row(row, "Asset Class"),
            "asset_class": safe_get_from_row(row, "Asset Class"),
            "asset_class_size": safe_get_from_row(row, "Management Style"),  # Using Management Style as proxy
            "asset_class_style": "",  # Not available in either source
            "general_region": safe_get_from_row(row, "Geographical Focus"),
            "specific_region": "",  # Not available

            # 3. Financial Metrics - PRIORITIZE sg_etfs_list
            "expense_ratio": safe_get_from_row(row, "TER (%)") or safe_get(info, "netExpenseRatio", ""),
            "aum": safe_get_from_row(row, "Val ($M)") or safe_get(info, "totalAssets", "") or safe_get(info, "netAssets", ""),
            "index_tracked": safe_get_from_row(row, "Underlying Benchmark"),

            # 4. Performance - PRIORITIZE sg_etfs_list
            "return_1m": safe_get_from_row(row, "TR 1M (%)"),
            "return_3m": safe_get_from_row(row, "TR 3M (%)") or safe_get(info, "trailingThreeMonthReturns", ""),
            "return_ytd": safe_get(info, "ytdReturn", ""),  # Not in sg_etfs_list
            "return_1y": safe_get_from_row(row, "Ann. TR 1Y (%)"),
            "return_3y": safe_get_from_row(row, "Ann. TR 3Y (%)") or safe_get(info, "threeYearAverageReturn", ""),
            "return_5y": safe_get(info, "fiveYearAverageReturn", ""),  # Not in sg_etfs_list

            # 5. Risk Metrics - yfinance only
            "52_week_low": safe_get(info, "fiftyTwoWeekLow", ""),
            "52_week_high": safe_get(info, "fiftyTwoWeekHigh", ""),
            "beta": safe_get(info, "beta3Year", ""),
            "std_dev": "",  # Not available

            # 6. Dividends - PRIORITIZE sg_etfs_list
            "annual_dividend_rate": safe_get_from_row(row, "Yield (%)") or safe_get(info, "yield", ""),
            "annual_dividend_yield": safe_get(info, "dividendYield", ""),
            "dividend": "",  # Not available
            "dividend_date": "",  # Not available

            # 7. Holdings - yfinance only
            "top_holdings": str(holdings_data) if holdings_data else "",
            "sector_breakdown": str(sector_data) if sector_data else "",
            
            # Metadata
            "source": "sg_etfs_list + yfinance",
            "last_updated": datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        }

        etf_metadata_list.append(data)
        
    except Exception as e:
        error_tickers.append((base_ticker, str(e)[:100]))
        print(f"\n❌ Error with {base_ticker}: {str(e)[:80]}")
    
    time.sleep(0.5)  # Rate limiting

# Create DataFrame
sg_etf_metadata_df = pd.DataFrame(etf_metadata_list)

# Save to CSV
output_path = "../data/official/sg/etf_metadata.csv"
sg_etf_metadata_df.to_csv(output_path, index=False)

print("\n" + "=" * 60)
print(f"✅ Successfully processed: {len(sg_etf_metadata_df)} ETFs")
print(f"❌ Errors encountered: {len(error_tickers)} ETFs")
print(f"✅ Saved to: {output_path}")

if error_tickers:
    print("\nFailed tickers:")
    for ticker, reason in error_tickers[:10]:  # Show first 10
        print(f"  - {ticker}: {reason}")

print("\n📊 Data Source Priority Summary:")
print("   Primary: sg_etfs_list (Trading Name, Fund Manager, TER, Performance, etc.)")
print("   Fallback: yfinance (Brand, Inception, YTD Return, Risk Metrics, Holdings)")

print("\nPreview:")
print(sg_etf_metadata_df[['ticker', 'etf_name', 'issuer', 'expense_ratio', 'return_1y']].head())
```
